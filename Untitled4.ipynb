```python
from sklearn.model_selection import train_test_split  # Para dividir os dados em conjuntos de treinamento e teste
from sklearn.linear_model import LinearRegression  # Para criar o modelo de regressão
from sklearn.metrics import mean_squared_error, r2_score  # Para avaliar o desempenho do modelo
```

```python
mkt.columns
```

```python
# Declarando variaveis

X = mkt[['youtube', 'facebook', 'newspaper']]
Y = mkt[['sales']]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 42)

print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Treinando o modelo

lm = LinearRegression()
lm.fit(X_train, Y_train)
```

```python
# Verificar a margem de erro do R2

y_pred = lm.predict(X_test)
r = r2_score(Y_test, y_pred)
print('r_quadrado:', r)
```

```python
c = [i for i in range (1, 53, 1)]
fig = plt.figure(figsize = (5,3))
plt.plot(c, Y_test, color = 'blue')
plt.plot(c, y_pred, color ='red')
plt.xlabel('index')
plt.ylabel('Vendas')
```
